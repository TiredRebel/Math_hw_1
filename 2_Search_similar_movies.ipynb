## Задача 2. Пошук схожих фільмів (косинусна подібність)

**Коментар:** Користувача та фільми представляємо як вектори інтересів/характеристик.  
Міра схожості — **косинусна подібність**:

$$
\cos\theta=\frac{\mathbf{u}\cdot\mathbf{v}}{\|\mathbf{u}\|\cdot\|\mathbf{v}\|}
$$

---

### Дано

$$
\mathbf{u}=(8,\,2,\,5)
$$

$$
\mathbf{v_A}=(9,\,1,\,2),\quad
\mathbf{v_B}=(1,\,9,\,8),\quad
\mathbf{v_C}=(7,\,2,\,6)
$$

---

## 1) Норма вектора користувача $\|\mathbf{u}\|$

**Коментар:** Норма (довжина) вектора: $\|\mathbf{x}\|=\sqrt{x_1^2+x_2^2+x_3^2}$.

$$
\|\mathbf{u}\|=\sqrt{8^2+2^2+5^2}=\sqrt{64+4+25}=\sqrt{93}
$$

---

## 2) Косинусна схожість з кожним фільмом

### 2.1) Фільм A

**Крок 1. Скалярний добуток $\mathbf{u}\cdot\mathbf{v_A}$:**

$$
\mathbf{u}\cdot\mathbf{v_A}=8\cdot9 + 2\cdot1 + 5\cdot2 = 72 + 2 + 10 = 84
$$

**Крок 2. Норма $\|\mathbf{v_A}\|$:**

$$
\|\mathbf{v_A}\|=\sqrt{9^2+1^2+2^2}=\sqrt{81+1+4}=\sqrt{86}
$$

**Крок 3. Косинусна подібність:**

$$
\cos(\mathbf{u},\mathbf{v_A})=\frac{84}{\sqrt{93}\cdot\sqrt{86}}
$$

---

### 2.2) Фільм B

**Крок 1. Скалярний добуток $\mathbf{u}\cdot\mathbf{v_B}$:**

$$
\mathbf{u}\cdot\mathbf{v_B}=8\cdot1 + 2\cdot9 + 5\cdot8 = 8 + 18 + 40 = 66
$$

**Крок 2. Норма $\|\mathbf{v_B}\|$:**

$$
\|\mathbf{v_B}\|=\sqrt{1^2+9^2+8^2}=\sqrt{1+81+64}=\sqrt{146}
$$

**Крок 3. Косинусна подібність:**

$$
\cos(\mathbf{u},\mathbf{v_B})=\frac{66}{\sqrt{93}\cdot\sqrt{146}}
$$

---

### 2.3) Фільм C

**Крок 1. Скалярний добуток $\mathbf{u}\cdot\mathbf{v_C}$:**

$$
\mathbf{u}\cdot\mathbf{v_C}=8\cdot7 + 2\cdot2 + 5\cdot6 = 56 + 4 + 30 = 90
$$

**Крок 2. Норма $\|\mathbf{v_C}\|$:**

$$
\|\mathbf{v_C}\|=\sqrt{7^2+2^2+6^2}=\sqrt{49+4+36}=\sqrt{89}
$$

**Крок 3. Косинусна подібність:**

$$
\cos(\mathbf{u},\mathbf{v_C})=\frac{90}{\sqrt{93}\cdot\sqrt{89}}
$$

---

## 3) Порівняння результатів і вибір найкращого фільму

**Коментар:** Найбільше значення косинусної подібності означає найбільшу схожість.

Оцінимо чисельно (приблизно):

$$
\cos(\mathbf{u},\mathbf{v_A})\approx 0.9393,\quad
\cos(\mathbf{u},\mathbf{v_B})\approx 0.5666,\quad
\cos(\mathbf{u},\mathbf{v_C})\approx 0.9894
$$

**Висновок:** найбільш схожий фільм — **Фільм C**, бо має найбільшу косинусну подібність.


```python
import numpy as np

# Вектори
u = np.array([8, 2, 5], dtype=float)
vA = np.array([9, 1, 2], dtype=float)
vB = np.array([1, 9, 8], dtype=float)
vC = np.array([7, 2, 6], dtype=float)

def cosine_similarity(a: np.ndarray, b: np.ndarray) -> float:
    # cos = (a·b) / (||a|| * ||b||)
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))

scores = {
    "Film A": cosine_similarity(u, vA),
    "Film B": cosine_similarity(u, vB),
    "Film C": cosine_similarity(u, vC),
}

for name, score in scores.items():
    print(f"{name}: {score:.6f}")

best = max(scores, key=scores.get)
print("\nBest match:", best)
```
