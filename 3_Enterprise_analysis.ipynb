## Задача 3. Аналіз виробництва (звіт)

**Коментар:** У виробництві витрати деталей за зміну можна описати системою лінійних рівнянь $A\mathbf{x}=\mathbf{b}$,  
де $\mathbf{x}$ — кількість зібраних дронів кожного типу.

---

### Дано

Нехай:
- $x_1$ — кількість дронів **"Розвідник"**
- $x_2$ — кількість дронів **"Камікадзе"**
- $x_3$ — кількість дронів **"Вантажний"**

Витрати на 1 дрон (мотори, контролери, акумулятори):

- "Розвідник": $(4,\ 1,\ 1)$  
- "Камікадзе": $(4,\ 1,\ 2)$  
- "Вантажний": $(6,\ 2,\ 4)$  

Загальні витрати за зміну:

$$
\mathbf{b}=
\begin{pmatrix}
460\\
130\\
240
\end{pmatrix}
$$

Формуємо матрицю коефіцієнтів $A$ (рядки — тип деталей, стовпці — тип дрона):

$$
A=
\begin{pmatrix}
4 & 4 & 6\\
1 & 1 & 2\\
1 & 2 & 4
\end{pmatrix},
\qquad
\mathbf{x}=
\begin{pmatrix}
x_1\\x_2\\x_3
\end{pmatrix}
$$

---

## 1) Система рівнянь

$$
\begin{cases}
4x_1 + 4x_2 + 6x_3 = 460 \\
x_1 + x_2 + 2x_3 = 130 \\
x_1 + 2x_2 + 4x_3 = 240
\end{cases}
$$

---

## 2) Перевірка єдиного розв’язку

**Коментар:** Якщо $\det(A)\neq 0$, система має **єдиний** розв’язок.

$$
\det(A)=
\begin{vmatrix}
4 & 4 & 6\\
1 & 1 & 2\\
1 & 2 & 4
\end{vmatrix}
=
4\begin{vmatrix}1 & 2\\2 & 4\end{vmatrix}
-4\begin{vmatrix}1 & 2\\1 & 4\end{vmatrix}
+6\begin{vmatrix}1 & 1\\1 & 2\end{vmatrix}
$$

Малі визначники:

$$
\begin{vmatrix}1 & 2\\2 & 4\end{vmatrix}=4-4=0,\quad
\begin{vmatrix}1 & 2\\1 & 4\end{vmatrix}=4-2=2,\quad
\begin{vmatrix}1 & 1\\1 & 2\end{vmatrix}=2-1=1
$$

Отже:

$$
\det(A)=4\cdot0-4\cdot2+6\cdot1=-2
$$

**Висновок:** $\det(A)=-2\neq 0$, тому розв’язок **існує і єдиний**.  
(Знак “-” лише означає зміну орієнтації базису, але на існування/єдиність розв’язку це не впливає.)

---

## 3) Аналітичний розв’язок з проміжними кроками

З другого рівняння:

$$
x_1 = 130 - x_2 - 2x_3
$$

Підставимо у третє:

$$
(130 - x_2 - 2x_3) + 2x_2 + 4x_3 = 240
$$

$$
130 + x_2 + 2x_3 = 240
\Rightarrow x_2 + 2x_3 = 110
\Rightarrow x_2 = 110 - 2x_3
$$

Повертаємось до $x_1$:

$$
x_1 = 130 - (110 - 2x_3) - 2x_3 = 20
$$

Підставимо в перше рівняння:

$$
4\cdot20 + 4x_2 + 6x_3 = 460
\Rightarrow 80 + 4x_2 + 6x_3 = 460
\Rightarrow 4x_2 + 6x_3 = 380
$$

Поділимо на 2:

$$
2x_2 + 3x_3 = 190
$$

Підставимо $x_2 = 110 - 2x_3$:

$$
2(110 - 2x_3) + 3x_3 = 190
\Rightarrow 220 - 4x_3 + 3x_3 = 190
\Rightarrow 220 - x_3 = 190
\Rightarrow x_3 = 30
$$

Тоді:

$$
x_2 = 110 - 2\cdot30 = 50
$$

---

## 4) Перевірка $A\mathbf{x}=\mathbf{b}$

$$
\mathbf{x}=
\begin{pmatrix}
20\\50\\30
\end{pmatrix}
$$

$$
A\mathbf{x}=
\begin{pmatrix}
4 & 4 & 6\\
1 & 1 & 2\\
1 & 2 & 4
\end{pmatrix}
\begin{pmatrix}
20\\50\\30
\end{pmatrix}
=
\begin{pmatrix}
4\cdot20+4\cdot50+6\cdot30\\
20+50+2\cdot30\\
20+2\cdot50+4\cdot30
\end{pmatrix}
=
\begin{pmatrix}
460\\130\\240
\end{pmatrix}
=\mathbf{b}
$$

---

## Відповідь

**Коментар:** За зміну було зібрано:

- **"Розвідник"** ($x_1$): **20**
- **"Камікадзе"** ($x_2$): **50**
- **"Вантажний"** ($x_3$): **30**


```python
import numpy as np

# A: рядки = (мотори, контролери, акумулятори), стовпці = (x1, x2, x3)
A = np.array([
    [4, 4, 6],
    [1, 1, 2],
    [1, 2, 4]
], dtype=float)

# b: загальні витрати деталей за зміну
b = np.array([460, 130, 240], dtype=float)

# 1) Визначник (перевірка єдиного розв'язку)
detA = np.linalg.det(A)
print("det(A) =", detA)

# 2) Розв'язок системи
x = np.linalg.solve(A, b)
print("x (float) =", x)

# 3) Перевірка Ax = b
check = A @ x
print("A @ x =", check)
print("b     =", b)

# 4) Зручний вивід як цілих (у цій задачі виходять цілі значення)
x_int = np.rint(x).astype(int)
print("\nКількість дронів за зміну:")
print(f'Розвідник (x1): {x_int[0]}')
print(f'Камікадзе (x2): {x_int[1]}')
print(f'Вантажний (x3): {x_int[2]}')
```
